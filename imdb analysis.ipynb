```python
import matplotlib.pyplot as plt

# Sample data
categories = ['Category A', 'Category B', 'Category C', 'Category D']
popularity = [20, 35, 42, 28]

# Create figure and axis objects
fig, ax1 = plt.subplots()

# Plot bar chart for popularity
ax1.bar(categories, popularity, color='blue')
ax1.set_xlabel('Market Category')
ax1.set_ylabel('Popularity')
ax1.set_title('Market Category vs. Popularity')

# Create secondary y-axis
ax2 = ax1.twinx()

# Plot line chart for another variable (if available)
# Example: Sales data
sales = [100, 150, 80, 120]
ax2.plot(categories, sales, color='red', marker='o')
ax2.set_ylabel('Sales')

# Display the combo chart
plt.show()


```
